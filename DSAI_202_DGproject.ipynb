```python
#clean all nulls in our data

# Calculate the mean of the 'cigsPerDay' column
mean_value_cigsPerDay = df['cigsPerDay'].mean()
# Fill NaN values in 'cigsPerDay' with the calculated mean and update the DataFrame in-place
df['cigsPerDay'].fillna(mean_value_cigsPerDay, inplace=True)


# Calculate the mean of the 'heartRate' column
mean_valueheartRate = df['heartRate'].mean()
# Fill NaN values in 'heartRate' with the calculated mean and update the DataFrame in-place
df['heartRate'].fillna(mean_valueheartRate, inplace=True)


# Calculate the mean of the 'glucose' column
mean_value_glucose = df['glucose'].mean()
# Fill NaN values in 'glucose' with the calculated mean and update the DataFrame in-place
df['glucose'].fillna(mean_value_glucose, inplace=True)


# Fill NaN values in 'currentSmoker' with 0 and update the DataFrame in-place
df['currentSmoker'].fillna(0, inplace=True)


# Find the last non-null value in the 'totChol' column to fill it in the all nulls
last_valid_value = df['totChol'].dropna().iloc[-1]
# Fill NaN values in 'totChol' with the last non-null value and update the DataFrame in-place
df['totChol'].fillna(last_valid_value, inplace=True)


# Find the last non-null value in the 'education' column to fill it in the all nulls
last_valid_value = df['education'].dropna().iloc[-1]
# Fill NaN values in 'education' with the last non-null value and update the DataFrame in-place
df['education'].fillna(last_valid_value, inplace=True)

# Find the last non-null value in the 'BPMeds' column to fill it in the all nulls
last_valid_value = df['BPMeds'].dropna().iloc[-1]
# Fill NaN values in 'BPMeds' with the last non-null value and update the DataFrame in-place
df['BPMeds'].fillna(last_valid_value, inplace=True)


# Calculate the mean of the 'BMI' column
mean_Number_BMI = df['BMI'].mean()
# Fill NaN values in 'BMI' with the calculated mean and update the DataFrame in-place
df['BMI'].fillna(mean_Number_BMI, inplace=True)
```

```python
#get and check permissions to make access control
class FileAccessControl:
    def __init__(self):
        self.permissions = {}

    def grant_permission(self, file, user, permission):
        if file not in self.permissions:
            self.permissions[file] = {}
        if user not in self.permissions[file]:
            self.permissions[file][user] = []
        self.permissions[file][user].append(permission)

    def check_permission(self, file, user, permission_requested):
        if file in self.permissions and user in self.permissions[file]:
            return permission_requested in self.permissions[file][user]
        return False
```

```python
file_access_control = FileAccessControl()
file_access_control.grant_permission("cleaned_data.csv", "user1", "read")
file_access_control.grant_permission("cleaned_data.csv", "user2", "read")
```

```python
#giving access to our cleaned data for user(i) with its access
def access_file(user, file, permission_requested):
    if file_access_control.check_permission(file, user, permission_requested):
        print(f"Access granted to {user} for {permission_requested} on {file}.")
    else:
        print(f"Access denied to {user} for {permission_requested} on {file}.")
```

```python
#giving permissions
access_file("user1", "cleaned_data.csv", "read")
access_file("user2", "cleaned_data.csv", "write")
access_file("user3", "cleaned_data.csv", "read")
```
